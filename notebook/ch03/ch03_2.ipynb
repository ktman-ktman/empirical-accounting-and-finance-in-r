# 3.2 for文の使い方

```julia
#ch03_14

r = 0.1
npv = -100 + 50 / (1 + r) + 50 / (1 + r)^2 + 50 / (1 + r)^3
println(npv)
```

```julia
#ch03_15
r = 0.1
npv = -100
for i in 1:3
    npv += 50 / (1 + r)^i
end
println(npv)
```

```julia
#ch03_16
r = 0.1
npv = -100
for i in 1:3
    println(npv)
    npv += 50 / (1 + r)^i
end
println(npv)
```

```julia
#ch03_17
if npv >= 0
    println("プロジェクトを実行")
end
```

```julia
#ch03_18
using Plots
r_l = range(0.1, 0.2, length=11)
npv_l = []
for (i, r_i) in enumerate(r_l)
    npv_i = -100
    for j in 1:3
        npv_i += 50 / (1 + r_i) ^ j
    end
    append!(npv_l, npv_i)
end

plot(
    r_l,
    npv_l,
    xlabel = "Discount Rate",
    ylabel = "NPV",
    title = "Figure: Discount Rate and NPV"
)
```

```julia
#ch03_19
r = 0.1
cf = [-100, 50, 50, 50]
years = 0:3
pv_cf = cf ./ (1 + r).^years
npv = sum(pv_cf)
println(npv)
```
