```julia
# ch03_01
r = 0.1
println(r)
```

```julia
# ch03_02
r = [
    0.1
    0.2
]
println(typeof(r))
println(r)
```

```julia
#ch03_03
r = range(0.1, 0.2, step=0.01)
r = collect(r)
println(typeof(r))
println(r)
```

```julia
# ch03_04
println(r[2])
i = 2
println(r[i])

println(r[2:5])

println(r[2:length(r)])
println(r[lastindex(r)])
```

```julia
#ch03_05
pv = [100 / (1 + r_i) for r_i in r]
println(pv)
```

```julia
#ch03_06
cf = collect(range(110, 120, 11))
println(cf)

if length(cf) == length(r)
    pv = zeros(0)
    for i in 1:length(cf)
        append!(pv, cf[i] / (1 + r[i]))
    end
    println(pv)
end
```

```julia
#ch03_07
using Plots
pv = [100 / (1 + r_i) for r_i in r]
plot(pv)
```

```julia
#ch03_08
plot(
    r,
    pv,
    xlabel="Discount Rate",
    ylabel="Present Value",
    title="Figure: Effector of Discount Rates on Present Value",
)
```

```julia
#ch03_09
plot(
    r,
    pv,
    xlabel="Discount Rate",
    ylabel="Present Value",
    title="Figure: Effector of Discount Rates on Present Value",
)
savefig("./PV.png")
savefig("./PV.pdf")
```
